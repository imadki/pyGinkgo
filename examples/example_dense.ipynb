## PyGinkgo ```dense.matrix``` Usage Example

This notebook demonstrates basic usage of PyGinkgo dense matrices:

- Initializing and filling a dense matrix using lists
- Creating dense matrix from 1D numpy array
- Creating 1D dense matrix from numpy array with stride
- Creating dense matrix from 2D numpy array
- Creating 1D dense matrix from a .mtx file
- SCaling and inverse scaling a dense matrix

````python
# importing necessary modules
import os
import numpy as np
import sys
sys.path.append("/Users/go/Code/pyGinkgo/build/Develop")

import pyGinkgo as pg
````

#### Creating dense matrix from 1D numpy array

````python
values = [1.0, 2.0, -1.0, 3.0, 4.0, -1.0, 5.0, 6.0, -1.0, 7.4]
ref = pg.ReferenceExecutor()
dense = pg.matrix.dense(ref, np.array(values))
for i in range(len(values)):
    print(dense.at(i))
````

#### Creating 1D dense matrix from numpy array with stride

````python
dense = pg.matrix.dense(ref, (3, 3), np.array([values]), 3)
for i in range(3):
    print("| ", end="")
    for j in range(3):
        print(dense.at(i, j), end=" ")
    print("|")
````

#### Creating dense matrix from 2D numpy array

````python
values = [[1.0, 2.0, -1.0, 3.0, 4.0], [-1.0, 5.0, 6.0, -1.0, 7.4]]
np_arr = np.array(values)
n_rows, n_columns = np_arr.shape
dense = pg.matrix.dense(ref, np.array(values))
for i in range(n_rows):
    print("| ", end="")
    for j in range(n_columns):
        print(dense.at(i, j), end=" ")
    print("|")
````

#### Creating 1D dense matrix from a .mtx file

````python
fn = os.path.join(os.getcwd(), "dense_example.mtx")
dense = pg.matrix.read_dense(fn, ref)
for i in range(12):
    print(dense.at(0, i))
````

#### Scaling and inverse scaling

````python
values = [1.0, 2.0, -1.0, 3.0, 4.0, -1.0, 5.0, 6.0, -1.0, 7.4]

dense = pg.matrix.dense(ref, (3, 3), np.array(values), 3)
# scaling
dense.scale(5)
print(dense.at(1))

# inverse scaling
dense.inv_scale(2)
print(dense.at(0))
````

````python
values = [1.0, 2.0, -1.0, 3.0, 4.0, -1.0, 5.0, 6.0, -1.0]
aT = np.array([values])
a = aT.T
dense_a = pg.matrix.dense(aT)
dense_b = pg.matrix.dense(a)
result = pg.matrix.dense(ref, (1, 1))
dense_a.apply(dense_b, result)
result.at(0) == aT.dot(a)[0, 0]
print(result.at(0))
````
